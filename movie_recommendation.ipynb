```python
import numpy as num
import pandas as pd
import difflib as dl
from sklearn.feature_extraction.text import TfidfVectorizer as tfv
from sklearn.metrics.pairwise import cosine_similarity

movies_set = pd.read_csv('/content/movies.csv')

selected_features = ['genres', 'tagline', 'keywords', 'cast', 'director']
print(selected_features)

for feature in selected_features:
  movies_set[feature] = movies_set[feature].fillna('')

combined_features = movies_set['genres'] + ' ' + movies_set['tagline'] + ' ' + movies_set['keywords'] + ' ' + movies_set['cast'] + ' ' + movies_set['director']
print(combined_features)
tfv = tfv()
feature_vectors = tfv.fit_transform(combined_features)
print(feature_vectors)
simi = cosine_similarity(feature_vectors)
print(simi.shape)
print(simi)
movie_name = input("Enter your favourite movie name: ")
title_list = movies_set['title'].tolist()
print(title_list)
close_match = dl.get_close_matches(movie_name,title_list)
print(close_match[0])
movie_index = movies_set[movies_set.title == close_match[0]]['index'].values[0]
print(movie_index)
simi_score = list(enumerate(simi[movie_index]))
print(simi_score)
sorted_simi = sorted(simi_score, key = lambda x:x[1], reverse = True)
print(sorted_simi)
print("movies suggestions:\n")
i = 1
for movie in sorted_simi:
  index = movie[0]
  index_title = movies_set[movies_set.index == index]['title'].values[0]
  if(i<21):
    print(i, '.', index_title)
    i+=1





```
